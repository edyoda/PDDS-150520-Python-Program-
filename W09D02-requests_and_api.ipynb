```python
## What is the difference an API or a normal Website?

API'S are generally used to exchange data.(JSON IS MOSTLY TO INTERCHANGE THE DATA WITH THE API)

1) CALL THE WAITER
2) YOU WILL TELL HIM THE ORDER DETAILS(LIKE PIZZA,COLD DRINKS)
3) WAITER WILL GO BACK IN THE KITCHEN AND TELL CHEF TO MAKE A PIZZA, AND PREPARE COLD DRINKS IN GLASSES.
4) WAITER WILL COME BACK AFTER SOME TIME WITH PREPARED FOOD (PIZZA AND COLD DRINK IN THIS CASE)

ANALOGY API = WAITER

1) CALL THE API
2) YOU WILL GIVE IT SOME INFO (PAYLOAD(IF IT IS A POST REQUEST) or PARAMS(IF IT IS A GET REQUEST))
3) API WILL GO THE SERVER SIDE LOGIC WHICH MIGHT ACCESS THE DATABASE AND PREPARE THE DATA WHICH WE HAVE REQUESTED.
4) API WILL COME BACK WITH THE INFO WHICH WE HAVE REQUESTED/OR RESPONSE TO THE REQUEST

NORMAL WEBSITES ARE FOR HUMAN READABLE AND ACCESIBLE.
BUT API'S FOR CODE TO READ AND ACCESS.

HOMEWORK: INSTALL POSTMAN, PLAY AROND THE API: "https://api.covid19api.com/summary"
```

```python
covid_response = requests.get("https://api.covid19api.com/summary")
```

```python
count = 0
for each_country in json_response["Countries"]:
    if each_country["NewConfirmed"] >= 40000:
        count+=1
        print(each_country["Country"])
print(count)


## Homework: to get some insights like this one that only india have more than 40000 cases confirmed in a single.
```

```python
count = 0
for each_country in json_response["Countries"]:
    if each_country["NewConfirmed"] - each_country["NewRecovered"] == 0 and each_country["NewConfirmed"]>0:
        print(each_country["NewConfirmed"],each_country["NewRecovered"])
        count+=1
        print(each_country["Country"])
print(count)
```
